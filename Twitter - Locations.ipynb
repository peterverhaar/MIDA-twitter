```python
import re
import os
import requests


dir = 'Twitter'

files = []

for file in os.listdir(dir):
    if re.search( r'txt$' , file  ):
        files.append(file)

for f in files:
    print(f)

    fh = open( os.path.join( dir, f )  , encoding = 'utf-8' )
    locations = dict()

    for line in fh:
        location = ''

        if re.search( r'^user-location' , line) and re.search(r'\w',line):
            location = re.sub( '^user-location:\s+' , '' , line  ).strip()
            #print(location)
            if re.search( r'\w' , location ):
                locations[location] = locations.get(location,0)+1

print( f'{len(locations)} locations. ')
```

```python
from urllib.parse import quote
import requests
import xml.etree.ElementTree as ET


def encode(text):
    text = re.sub(r'[^\x00-\x7F]+',' ', text )
    text = re.sub( r'\d' , '' , text)
    text = re.sub( r'[#]' , ''  , text )
    text = quote(text)
    return text

coord = dict()

for i,l in enumerate(locations):
    print(f'{i}. {l}')
    l = encode(l)
    url = 'https://nominatim.openstreetmap.org/search?q='+ l + '&format=xml'


    print("Trying {}".format(url))
    response = requests.get( url )
    
    # Parse the XML using the ElementTree library and find places
    root = ET.fromstring( response.text )
    places = root.findall('place')
    if len(places) > 0:
        place = places[0]
        lat = place.attrib['lat']
        lon = place.attrib['lon']
        coord[l] = (lat,lon)

```

```python
import json

countries = dict()

for l in coord:
    lat = coord[l][0]
    lon = coord[l][1]
        
    url = f"http://api.geonames.org/countryCodeJSON?lat={lat}&lng={lon}&username=peter.verhaar"
    response = requests.get( url )
    if response:
        geo_data = response.json()
        #print(tajson_str)
        #geo_data = json.loads(json_str)
        try: 
            countries[l] = (geo_data["countryCode"],geo_data["countryName"])
        except:
            print( f'Skipped {url}' )
        
```

```python
out = open( 'countries.csv' , 'w' , encoding = 'utf-8' )

out.write('location,lat,long,code,country\n')

for l in coord:
    out.write( f'{l},{coord[l][0]},{coord[l][0]},' )
    if l in countries:
        country_code = countries[l][0]
        country_name = countries[l][1]
        out.write(f'{country_code},{country_name}')
    else:
        out.write(',')
    out.write('\n')
    
out.close()
```
